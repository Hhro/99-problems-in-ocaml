# Working with lists

## 1. Write a function last : 'a list -> 'a option that returns the last element of a list

```ocaml
let rec last = function
    | [] -> None
    | [x] -> Some x
    | _ :: t -> last t;;

last ["a"; "b"; "c"; "d"];;
last [];;
```

```ocaml
let rec last_two = function
    | [] | [_] -> None
    | [x; y] -> Some (x, y)
    | h :: t -> last_two t;;

last_two ["a"; "b"; "c"; "d"];;
last_two ["a"];;
```

```ocaml
let rec at k = function
    | [] -> None
    | h :: t -> if k=1 then Some h else at (k-1) t;;

at 3 ["a"; "b"; "c"; "d"; "e"];;
    
```

```ocaml
let length list =
    let rec aux n = function
        | [] -> n
        | _::t -> aux (n+1) t
    in aux 0 list;;
    
length ["a"; "b"; "c"];;
length [];;
```

```ocaml
let rev list = 
    let rec aux acc = function
        | [] -> acc
        | h :: t -> aux (h :: acc) t in
    aux [] list;;

rev ["a"; "b"; "c"];;
```

```ocaml
let is_palindrome list=  
    rev list = list;;

is_palindrome ["x"; "a"; "m"; "a"; "x"];;
not (is_palindrome ["a"; "b"]);;
```

```ocaml
type 'a node = 
    | One of 'a
    | Many of 'a node list;;

let flatten list = 
    let rec aux acc = function
        | [] -> acc
        | One x :: t -> aux (x::acc) t
        | Many l :: t -> aux (aux acc l) t in
    rev (aux [] list);;

flatten [One "a"; Many [One "b"; Many [One "c" ;One "d"]; One "e"]];;
```

```ocaml
let rec compress = function 
    | x :: (y :: _ as t) -> if x=y then compress t else x :: compress t
    | z -> z;;

compress ["a"; "a"; "a"; "a"; "b"; "c"; "c"; "a"; "a"; "d"; "e"; "e"; "e"; "e"];;
```

```ocaml
let pack list =
    let rec aux current acc = function
        | [] -> []
        | [x] -> (x::current) :: acc
        | x :: (y :: _ as t) -> 
            if x=y then aux (x::current) acc t
            else aux [] ((x::current) :: acc) t in
    rev (aux [] [] list);;

pack ["a"; "a"; "a"; "a"; "b"; "c"; "c"; "a"; "a"; "d"; "d"; "e"; "e"; "e"; "e"];;
```
